Q1. Write a unique paragraph (5-6 sentences) about your favorite topic (e.g., sports, technology, food, books, etc.).

1.Convert text to lowercase and remove punctuaƟon.
2.Tokenize the text into words and sentences.
3.Remove stopwords (using NLTK's stopwords list).
4.Display word frequency distribuƟon (excluding stopwords).

```python
import string

text = """Books have always been a source of comfort and inspiration for me. I love how a good story can transport you to another world, introduce you to unforgettable characters, and make you feel a wide range of emotions. Whether it's fantasy, mystery, or historical fiction, there's something magical about turning the pages and not knowing what's coming next. Reading also sparks my imagination and helps me understand different perspectives. It's more than just a hobby—it’s a way to grow, escape, and connect with others through shared stories."""
# Convert to lowercase
text_lower = text.lower()

# Remove punctuation
text_no_punct = text_lower.translate(str.maketrans('', '', string.punctuation))

print(text_no_punct)

     
```

```python
import nltk
nltk.download('punkt')

     
```

```python
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import word_tokenize, sent_tokenize

# Tokenize into words
words = word_tokenize(text_no_punct)

# Tokenize into sentences
sentences = sent_tokenize(text_no_punct)

print(words)
print(sentences)
```

```python
nltk.download('stopwords')
from nltk.corpus import stopwords

# Remove stopwords
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words if word not in stop_words]

print(filtered_words)
```

```python
from nltk.probability import FreqDist

# Frequency distribution
fdist = FreqDist(filtered_words)

# Display the frequency distribution
fdist.plot(20, title="Word Frequency Distribution")
```

Q2: Stemming and Lemmatization

1.Take the tokenized words from Question 1 (after stopword removal).
2.Apply stemming using NLTK's PorterStemmer and LancasterStemmer.
3.Apply lemmatization using NLTK's WordNetLemmatizer.
4.Compare and display results of both techniques.

```python
filtered_words
```

```python
from nltk.stem import PorterStemmer,LancasterStemmer

ps=PorterStemmer()
ls=LancasterStemmer()

porter_stemmed=[ps.stem(word) for word in filtered_words]
lancaster_stemmed=[ls.stem(word) for word in filtered_words]

print(porter_stemmed)
print(lancaster_stemmed)
```

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')

lemmatizer=WordNetLemmatizer()

lemmatized_words=[lemmatizer.lemmatize(word) for word in filtered_words]

print(lemmatized_words)
     
```

```python
text = "Books have always been a source of comfort and inspiration for me. I love how a good story can transport you to another world, introduce you to unforgettable characters, and make you feel a wide range of emotions. Whether it's fantasy, mystery, or historical fiction, there's something magical about turning the pages and not knowing what's coming next. Reading also sparks my imagination and helps me understand different perspectives. It's more than just a hobby—it’s a way to grow, escape, and connect with others through shared stories."
```

```python
words_more_than_5 = re.findall(r'\b\w{6,}\b', text)
print("Words with more than 5 letters:", words_more_than_5)
```

```python
words_starting_vowel = re.findall(r'\b[AEIOUaeiou][a-zA-Z]*\b', text)
print("Words starting with a vowel:", words_starting_vowel)
```

Q4. Custom Tokenization & Regex-based Text Cleaning

1.Take original text from Question 1.
2.Write a custom tokenization function that: a. Removes punctuation and special symbols, but keeps contractions (e.g., "isn't" should not be split into "is" and "n't"). b. Handles hyphenated words as a single token (e.g., "state-of-the-art" remains a single token). c. Tokenizes numbers separately but keeps decimal numbers intact (e.g., "3.14" should remain as is).
3.Use Regex Substitutions (re.sub) to: a. Replace email addresses with '' placeholder. b. Replace URLs with '' placeholder. c. Replace phone numbers (formats: 123-456-7890 or +91 9876543210) with '' placeholder.

```python
import re

text = """Books have always been a source of comfort and inspiration for me. I love how a good story can transport you to another world.
Visit https://example.com for more details or contact me at john.doe123@example.com or +91 9876784949."""

# Step 1: Custom Tokenization Function
def custom_tokenizer(text):
    # Remove punctuation except hyphens, apostrophes (for contractions), and decimals
    text = re.sub(r"[^\w\s\.\-']", '', text)

    # Split into tokens
    tokens = text.split()

    return tokens

# Step 2: Regex Substitution Function
def clean_text(text):
    # Replace email addresses with 
    text = re.sub(r'\b[\w\.-]+@[\w\.-]+\.\w{2,4}\b', '', text)

    # Replace URLs with 
    text = re.sub(r'https?://\S+|www\.\S+', '', text)

    # Replace phone numbers with 
    text = re.sub(r'(\+?\d{1,3}[-\s]?\d{10}|\d{3}-\d{3}-\d{4})', '', text)

    return text

# Step 3: Run the functions
# 3a. First clean the text
cleaned_text = clean_text(text)
print("Cleaned Text:\n", cleaned_text)

# 3b. Then tokenize the cleaned text
tokens = custom_tokenizer(cleaned_text)
print("\nCustom Tokens:\n", tokens)

     
```
