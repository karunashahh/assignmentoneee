Q1. Ask the user to enter a value (e.g., M) for a mathematical function.  Generate x values from -10 to 10 using np.linspace().  Compute y values for: o y= M⋅x 2 o y=M⋅sin(x)  Plot both functions in a single figure: o Use different colors and line styles. o Add legend, grid, and title.

```python
import numpy as np
import matplotlib.pyplot as plt

n = float(input("Enter a value for n: "))

x = np.linspace(-10, 10, 100)

y1 = n * x**2
y2 = n * np.sin(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y1, color='blue', linestyle='-', label=f'y = {n}x²')
plt.plot(x, y2, color='red', linestyle='--', label=f'y = {n}sin(x)')

plt.ylim(-5, 50);

plt.legend()
plt.grid(True)
plt.title(f'Plots of y = {n}x² and y = {n}sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Q.2 Create a dataset of five subjects and scores.  Convert it into a Pandas DataFrame.  Plot the scores using a Seaborn bar plot with: o Different colors for each bar. o Annotations on top of each bar. o Title, axis labels, and grid.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create a dataset of subjects and scores
data = {
    'Subject': ['english', 'Science', 'maths', 'History', 'computer'],
    'Score': [76, 90, 78, 80, 88]
}

# Convert to Pandas DataFrame
df = pd.DataFrame(data)

# Plot scores using a Seaborn bar plot
plt.figure(figsize=(8, 5))
colors = sns.color_palette('husl', len(df))
plot = sns.barplot(x='Subject', y='Score', data=df, palette=colors)

# Add annotations on top of each bar
for index, row in enumerate(df.iterrows()):
    plt.text(index, row[1].Score + 1, str(row[1].Score), ha='center')

# Add title, axis labels, and grid
plt.title('Scores by Subject')
plt.xlabel('Subject')
plt.ylabel('Score')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

Q3. Write a Python script to select your roll number as seed for NumPy and then generate a dataset of 50 values using np.random.randn().  Create a 2x2 subplot layout displaying: o Line plot showing cumulative sum. o Scatter plot with random noise.  Customize with titles, axis labels, and grid

```python
import numpy as np
import matplotlib.pyplot as plt

# Select roll number as seed for NumPy (replace 102346 with your roll number)
np.random.seed(102346)

# Generate a dataset of 50 values using np.random.randn()
data = np.random.randn(50)

# Create a 2x2 subplot layout
plt.figure(figsize=(10, 6))

# Line plot showing cumulative sum
plt.subplot(2, 2, 1)
plt.plot(np.cumsum(data), color='blue')
plt.title('Cumulative Sum')
plt.xlabel('Index')
plt.ylabel('Cumulative Value')
plt.grid(True)

# Scatter plot with random noise
plt.subplot(2, 2, 2)
x = np.arange(50)
y = data + np.random.normal(scale=0.5, size=50)
plt.scatter(x, y, color='red')
plt.title('Scatter Plot with Noise')
plt.xlabel('Index')
plt.ylabel('Value')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
Q.4 Download Data-set from the below link https://github.com/AnjulaMehto/MCA/blob/main/company_sales_data.csv Apply ‘seaborn’ library to do the following.

1.Read Total profit of all months and show it using a line plot.
2.Read all product sales data and show it using a multiline plot.
3.Plot bar chart for all the features/attributes.
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Download and read the dataset
dataset_url = 'https://github.com/AnjulaMehto/MCA/raw/main/company_sales_data.csv'
df = pd.read_csv(dataset_url)

# 1. Line plot for total profit of all months
plt.figure(figsize=(9, 6))
sns.lineplot(x=df['month_number'], y=df['total_profit'], marker='o', color='blue')
plt.title('Total Profit Over Months')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.grid(True)
plt.show()

# 2. Multiline plot for all product sales data
plt.figure(figsize=(9, 6))
for column in df.columns[1:-1]:  # Exclude month_number and total_profit
    sns.lineplot(x=df['month_number'], y=df[column], label=column)
plt.title('Product Sales Over Months')
plt.xlabel('Month Number')
plt.ylabel('Sales')
plt.grid(True)
plt.legend()
plt.show()

# 3. Bar chart for all features/attributes
plt.figure(figsize=(9, 6))
df.sum().drop('month_number').plot(kind='bar', color='purple')
plt.title('Total Sales and Profit by Feature')
plt.xlabel('Features')
plt.ylabel('Total Values')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
